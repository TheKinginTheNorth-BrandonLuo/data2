```python
import numpy as np
import pandas as pd
ratings = pd.read_csv('ratings.csv', usecols = ['userId', 'movieId','rating'])
movies = pd.read_csv('movies.csv')
data = pd.merge(movies, ratings)
data.head(5)
```

```python
data_rating = data.groupby('title').agg({'rating':[np.size,np.mean]})

rating_mean_movie = pd.merge(movies, data_rating, left_on ='title', right_on = 'title')

rating_mean_movie.to_csv('rating_mean.csv')
rating_mean_movie.head(5)
```

```python
rating_mean_movie=rating_mean_movie.rename(columns = {"(rating, size)":"rating_size"})
```
