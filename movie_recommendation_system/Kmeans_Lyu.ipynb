```python
def clustering_errors(k,data):
    kmeans = KMeans(n_clusters=k).fit(data)
    predictions = kmeans.predict(data)
    sc_score = silhouette_score(data, predictions)
    return sc_score
```

```python
# genre_a = csr_matrix(genre)
# genre_b = pd.SparseDataFrame(genre_a)
# genre_C = genre_b.to_coo()
from sklearn.cluster import KMeans
from scipy.sparse import csr_matrix
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
D = {}
# genre_C = csr_matrix(pd.SparseDataFrame(genre).to_coo(sort_labels = bool))
kmeans_1 = KMeans(n_clusters = 20, max_iter = 500, n_init = 10, algorithm = 'full').fit(genre)
print(kmeans_1.labels_)
for i,l in enumerate(kmeans_1.labels_):
    D[genre.index[i]] = l

import csv
with open('order.csv','w') as f:
    writer = csv.writer(f)
    for key, value in D.items():
        writer.writerow([key, value])
        
```

```python
cluster_data = pd.read_csv('order.csv', header = None, names = ['title', 'cluster'])
cluster_rating = pd.read_csv('rating_mean.csv', usecols = [1,2,3,5])
cluster_rating.columns = ['movieId', 'title', 'genres', 'rating_mean']
cluster_rating_data = pd.merge(cluster_rating, cluster_data)

def recommendation(pred):
    cluster = cluster_rating_data.loc[cluster_rating_data['cluster']==pred]
    cluster.sort_values(by = ['rating_mean'], ascending = False).to_csv('predicted_cluster.csv')
    pred_cluster = pd.read_csv('predicted_cluster.csv', usecols = ['title','genres','rating_mean'])
    return pred_cluster.loc[0:9]

cluster_rating_data.head(20)
```

```python
import numpy as np
from sklearn.decomposition import PCA
reduced_data =PCA(n_components=2).fit_transform(genre)
kmeans = KMeans(init = 'k-means++', n_clusters=20, n_init=10)
kmeans.fit(reduced_data)

h = .02     
x_min, x_max = reduced_data[:, 0].min() - 0.2, reduced_data[:, 0].max() + 0.2
y_min, y_max = reduced_data[:, 1].min() - 0.2, reduced_data[:, 1].max() + 0.2
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
 
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10,6))
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')
 
plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on the movies dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```
