```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('data2.txt',header = None,names=['X','Y'])
fig,axes = plt.subplots()
data.plot(kind='scatter',x='X',y='Y',ax=axes,marker='o',color='r')
axes.set(xlabel='X',ylabel='Y')
data.insert(0,'Ones',1)#Add a column to the data to facilitate matrix calculation
data.head()
# Define extraction functions for data features and tags
def get_fea_lab(data):
    cols = data.shape[1]
    x = data.iloc[:,0:cols-1] # X is the first two columns in data (excluding index columns)
    y = data.iloc[:,cols-1:cols] # Y is the last column in data
 
    # Convert both X and y into matrix form
    x = np.matrix(x.values)
    y = np.matrix(y.values)
    return x,y
 
# Define the cost function:
def computeCost(data,theta,i):
    x,y = get_fea_lab(data)
    inner = np.power(((x*theta.T)-y),2)
    return (float(inner[i]/2))
 
# Define the stochastic gradient descent function
def stochastic_gradient_descent(data,theta,alpha,epoch):
    
    x0,y0 = get_fea_lab(data)  #  take the X and y matrices
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.shape[1])
    cost = np.zeros(len(x0))
    avg_cost = np.zeros(epoch)
    
    for k in range(epoch):
        new_data = data.sample(frac=1) # Disrupted data
        x,y = get_fea_lab(new_data)  # take the new X and y matrices
        
        for i in range(len(x)):
            error = x[i:i+1]*theta.T-y[i]
            cost[i] = computeCost(new_data,theta,i)
        
            for j in range(parameters):
                temp[0,j] = theta[0,j] - alpha*error*x[i:i+1,j]
            
            theta = temp
        avg_cost[k] = np.average(cost)
        
    return theta,avg_cost
 
# Initialization learning rate, iteration round and parameter theta
alpha = 0.001
epoch = 1500
theta = np.matrix(np.array([0,0]))
 
# The random gradient descent function is called to calculate the theat parameter in linear regression
g,avg_cost = stochastic_gradient_descent(data,theta,alpha,epoch)

fig, axes = plt.subplots()
axes.plot(np.arange(epoch), avg_cost, 'r')
axes.set_xlabel('Iterations')
axes.set_ylabel('cost')
axes.set_title('cost vs. Iterations')

x = np.linspace(data.X.min(),data.Y.max(),97)
f = g[0,0] + g[0,1]*x
 
fig,axes = plt.subplots()
axes.plot(x,f,'r',label='Fitted')
axes.scatter(x=data.X,y=data.Y,label='Trainning data')
axes.legend(loc='best')
axes.set(xlabel='X',ylabel='Y',title='X vs. Y')
plt.show()
```
